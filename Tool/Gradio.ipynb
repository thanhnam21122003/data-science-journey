```python
import gradio as gr

def greet(name, intensity):
  return "Hello, " + name + "!" * int(intensity)

demo = gr.Interface(
    fn = greet, # hàm Python, thể hiện logic của chương trình
    inputs = ["text", "slider"], # danh sách các component có sẵn cho phần input
    outputs = ["text"], # danh sách các component có sẵn cho phần output
    api_name = "predict"
)

demo.launch()
```

```python
def display_data(text, img_path, slider_val, choices_list):
  return {
      "1. Text": text,
      "2. Image Path": img_path,
      "3. Slider Value": slider_val,
      "4. Choices": choices_list,
  }
```

```python
my_inputs = [
    gr.Textbox(label = "Type something"),
    gr.Image(type = "filepath", label = "Upload Image"),
    gr.Slider(minimum = 1, maximum = 10, step = 1, label = "Pick a number"),
    gr.CheckboxGroup(choices = ["Option A", "Option B", "Option C"], label = "Select Options"),
]

demo = gr.Interface(
    fn = display_data,
    inputs = my_inputs,
    outputs = "json",
)

demo.launch()
```

* **gr.Blocks()**: khối nền tảng dùng để xây dựng giao diện Gradio một cách linh hoạt, cho phép người dùng tổ chức và kết hợp các component cũng như các layout con như Row, Column, Tab
* **gr.Row()**: dùng để sắp xếp các component theo chiều ngang, từ trái sang phải, trong cùng một hàng
* **gr.Column()**: dùng để sắp xếp các component theo chiều dọc bên trong một cột, thường được kết hợp với **gr.Row()** để tạo bố cục nhiều cột
* **gr.Tab()**: dùng để giao diện thành nhiều tab khác nhau, giúp tổ chức nội dung gọn gàng và thuận tiện hơn cho người dùng khi tương tác

```python
with gr.Blocks() as demo:
  with gr.Tabs():
    # TAB 1: Demonstrating Row (Horizontal)
    with gr.Tab("Tab 1: Horizontal Layout"):
      with gr.Row():
        gr.Textbox(label = "Input A")
        gr.Textbox(label = "Input B")
        gr.Button("Submit")
    with gr.Tab("Tab 2: Vertical Layout"):
      with gr.Row():
        # Left Column (Input Area)
        with gr.Column():
          gr.Textbox(label = "User Name")
          gr.Slider(label = "Age", maximum = 100)
          gr.Checkbox(label = "Subcribe to newsletter")
          gr.Button("Sign Up")
        # Right Column (Output/ Preview Area)
        with gr.Column():
          gr.Image(label = "Profile Preview")
          gr.JSON(label = "Raw Data")
demo.launch()
```

Một số event listener phổ biến bao gồm:
* **.click()**: khi người dùng nhấn vào một nút (Button)
* **.submit()**: khi người dùng nhấn phím Enter trong các component nhập liệu như Textbox
* **.change()**: khi giá trị của một component đầu vào thay đổi, ví dụ như Slider hoặc Checkbox
* **.input()**: listen mọi thay đổi đầu vào theo thời gian thực, thường dùng trong các ứng dụng cần phản hồi liên tục

```python
def welcome(name):
  return f"Welcome, {name}!"

with gr.Blocks() as demo:
  gr.Markdown("### Event Listener Demo")

  # Example 1: Button Click (Manual)
  with gr.Row():
    in_1 = gr.Textbox(label = "Enter Name (Manual)")
    btn = gr.Button("Update")
    out_1 = gr.Textbox(label = "Result")
  # Listener 1: Only runs when Button is clicked
  btn.click(fn = welcome, inputs = in_1, outputs = out_1)

  # Example 2: Change Event (Real-time)
  with gr.Row():
    in_2 = gr.Textbox(label = "Enter Name (Real-time)")
    out_2 = gr.Textbox(label = "Result")
  # Listener 2: Runs every time you type a character
  in_2.change(fn = welcome, inputs = in_2, outputs = out_2)

demo.launch()
```
